```python
bpm_pace_dist = df[['Avstand', 'Gjennomsnittlig puls', 'Tid', 'Dato', 'Makspuls']]
```

```python
from datetime import datetime, timedelta

def time_to_seconds(timestamp):
    t = datetime.strptime(timestamp[:8], "%H:%M:%S")
    delta = timedelta(hours=t.hour, minutes=t.minute, seconds=t.second)
    return int(delta.total_seconds())
    
bpm_pace_dist['Tid_in_seconds'] = bpm_pace_dist['Tid'].apply(lambda d: str(time_to_seconds(d)))
```

```python
bpm_pace_dist
```

```python
def format_pace(df):
    m, s = divmod(int(df['Pace_in_seconds_per_km']), 60)
    c = "0"+str(int(s)) if len(str(int(s))) == 1 else str(int(s))
    return f'{int(m)}:{c}'

def calculate_pace(df):
    return int(float(df['Tid_in_seconds']) / float(df['Avstand']))

bpm_pace_dist['Pace_in_seconds_per_km'] = bpm_pace_dist.apply(lambda d: str(calculate_pace(d)), axis=1)
bpm_pace_dist['Pace_formatted'] = bpm_pace_dist.apply(lambda d: str(format_pace(d)), axis=1)
```

```python
bpm_pace_dist
```

```python
def pace_to_speed(df):
    return 3600 / int(df['Pace_in_seconds_per_km'])

bpm_pace_dist['Average_speed'] = bpm_pace_dist.apply(lambda x: pace_to_speed(x), axis=1)
```

```python
d = bpm_pace_dist.astype({"Avstand": float})
d['Dato'] =  pd.to_datetime(d['Dato'])

bpm_pace_dist = d[d['Avstand'] >= 5]
```
